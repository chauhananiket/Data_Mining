```python
import numpy as np
import pandas as pd
import json
import copy
from tabulate import tabulate
import matplotlib.pyplot as plt
from datetime import date
```

# Preparing Dataset

```python
# Opening JSON file
f = open('/content/drive/MyDrive/DMG_Assignment/Assignment_0/states_daily.json')
# f = open('/content/drive/MyDrive/Projects/DMG_Assignment/Assignment_0/states_daily.json')

# returns JSON object as
# a dictionary
data = json.load(f)

# Closing file
f.close()
```

```python
Dict = data['states_daily']
col = list(Dict[0].keys())
dataframe = pd.DataFrame(columns=col)
state_list = copy.deepcopy(col)
state_list.remove('date')
state_list.remove('dateymd')
state_list.remove('status')
state_list.remove('tt')
state_list.remove('un')
print(dataframe.columns)
for i in range(len(Dict)):
  dataframe.loc[len(dataframe.index)] = list(Dict[i].values())
```

```python
df_confirmed = pd.DataFrame(columns=list(Dict[0].keys()))
df_recovered = pd.DataFrame(columns=list(Dict[0].keys()))
df_deceased = pd.DataFrame(columns=list(Dict[0].keys()))
for i in range(len(Dict)):
  if dataframe.iloc[i]['status'] == "Confirmed":
    df_confirmed.loc[len(df_confirmed.index)] = list(Dict[i].values())
  elif dataframe.iloc[i]['status'] == "Recovered":
    df_recovered.loc[len(df_recovered.index)] = list(Dict[i].values())
  elif dataframe.iloc[i]['status'] == "Deceased":
    df_deceased.loc[len(df_deceased.index)] = list(Dict[i].values())
```

```python
for i in range(len(col)):
  if col[i] != 'date' and col[i] != 'dateymd' and col[i] != 'status':
    dataframe[col[i]] = dataframe[col[i]].astype(int)
    df_recovered[col[i]] = df_recovered[col[i]].astype(int)
    df_confirmed[col[i]] = df_confirmed[col[i]].astype(int)
    df_deceased[col[i]] = df_deceased[col[i]].astype(int)
```

```python
df_confirmed_del = df_confirmed.drop(['date','dateymd','status','tt','un'], axis = 1)
df_recovered_del = df_recovered.drop(['date','dateymd','status','tt','un'], axis = 1)
df_deceased_del = df_deceased.drop(['date','dateymd','status','tt','un'], axis = 1)
```

```python
states_only = ['ap', 'ar', 'as', 'br','ct', 'ga', 'gj', 'hp', 'hr', 'ka', 'kl', 'jh', 'mh', 'ml', 'mn', 'mp',
       'mz', 'nl', 'or', 'pb', 'rj', 'sk', 'tg', 'tn', 'tr', 'up', 'ut','wb']
ut_only = ['an', 'ch', 'dd', 'dl', 'dn', 'jk', 'la', 'ld', 'py']
```

## Q6 - Find the day and count with the highest spike in a day in the number of cases for each state and UTs for “Confirmed”, “Recovered” and “Deceased” between dates 14-Mar-2020 and 16-Aug-2021.

```python
q6_df_c = pd.DataFrame(columns=['State', 'Highest Number of Confirmed cases', 'Date'])
q6_df_r = pd.DataFrame(columns=['State', 'Highest Number of Recovered cases', 'Date'])
q6_df_d = pd.DataFrame(columns=['State', 'Highest Number of Deceased cases', 'Date'])
for i in range(len(state_list)):
  ans1 = [state_list[i], df_confirmed[state_list[i]].max(), df_confirmed['date'][df_confirmed[state_list[i]].argmax()]]
  q6_df_c.loc[len(q6_df_c)] = ans1
  ans2 = [state_list[i], df_recovered[state_list[i]].max(), df_recovered['date'][df_recovered[state_list[i]].argmax()]]
  q6_df_r.loc[len(q6_df_r)] = ans2
  ans3 = [state_list[i], df_deceased[state_list[i]].max(), df_deceased['date'][df_deceased[state_list[i]].argmax()]]
  q6_df_d.loc[len(q6_df_d)] = ans3
print(tabulate(q6_df_c, headers = ['State', 'Highest Number of Confirmed cases', 'Date'], tablefmt = 'psql'))
print(tabulate(q6_df_c, headers = ['State', 'Highest Number of Recovered cases', 'Date'], tablefmt = 'psql'))
print(tabulate(q6_df_c, headers = ['State', 'Highest Number of Deceased cases', 'Date'], tablefmt = 'psql'))
```

## Q1 - Plot the area trend line for total “Confirmed”, “Recovered” and “Deceased” cases from 14-Mar-2020 to 16-Aug-2021.


```python
other = [i for i in list(dataframe.columns) if i not in ['date','dateymd','status','tt', 'un']]

confirmarray = np.sum(np.array(df_confirmed[other]),axis=1)
recoveredarray = np.sum(np.array(df_recovered[other]),axis=1)
deceasedarray = np.sum(np.array(df_deceased[other]),axis=1)

df_plot = pd.DataFrame()
df_plot['Confirm'] = list(confirmarray)
df_plot['Recover'] = list(recoveredarray)
df_plot['Decrease'] = list(deceasedarray)

df_plot.plot.area(stacked=False, figsize=(15,5));
plt.ylabel("Count")
plt.xlabel("Dates")
plt.show()
```

## Q3 - Plot the area trend line for active cases. Assume active = Confirmed - (Recovered + Deceased) from 14-Mar-2020 to 16-Aug-2021.

```python
other = [i for i in list(dataframe.columns) if i not in ['date','dateymd','status','tt', 'un']]

confirmarray = np.sum(np.array(df_confirmed[other]),axis=1)
recoveredarray = np.sum(np.array(df_recovered[other]),axis=1)
deceasedarray = np.sum(np.array(df_deceased[other]),axis=1)
activearray = confirmarray - (recoveredarray + deceasedarray)
df_active = pd.DataFrame()
df_active['active'] = list(activearray)
df_active.plot.area(stacked=False,
                  figsize=(15,5));
plt.show()
```

## Q4 - Plot a bar plot of the number of active cases in Delhi, Tamil Nadu and Gujarat for any date range of your choice.

```python
#@markdown Input of Date Format : Year-Month-Day
Start_Date = "2020-03-14" #@param {type:"string"}
End_Date =  "2020-05-16"#@param {type:"string"}
```

```python
indexs = df_confirmed.index[df_confirmed['dateymd'] == Start_Date].tolist()[0]
indexe = df_confirmed.index[df_confirmed['dateymd'] == End_Date].tolist()[0]
getrange = [i for i in range(indexs,indexe+1)]
df_confirmed_range = np.array(df_confirmed[['dl','tn','gj']].iloc[getrange])
df_recovered_range = np.array(df_recovered[['dl','tn','gj']].iloc[getrange])
df_decreased_range = np.array(df_deceased[['dl','tn','gj']].iloc[getrange])
df_active_range = df_confirmed_range - (df_recovered_range + df_decreased_range)
df_active_range_sum = np.sum(df_active_range,axis=0)
plot_data = {'Delhi':df_active_range_sum[0],'Tamil Nadu':df_active_range_sum[1],'Gujarat':df_active_range_sum[2]}

fig = plt.figure(figsize = (11, 6))
plt.bar(plot_data.keys(),plot_data.values(),width = 0.4)
plt.xlabel("States")
plt.ylabel("Active Cases in date range")
plt.show()
```
